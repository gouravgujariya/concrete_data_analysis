## feature analysis

### Age vs Strength

The age of concrete structures has a lot to do with their strength and durability properties. Understanding the strength-time relationship of concrete helps to know the effect of loading at a later age. 

**From the below graph**, we can say that the max strength of the mixture is when is between **40 to 90**. Age has strength max out and as the **age has increased, it decreases the strength**

```python
sns.scatterplot(df['Age'],df['Strength'])
```

### cement vs strength

Cement and strength are directly proportionate because as the cement concentration in the mixture is increased, it increased the strength but if concentration increased more than required will result in resources wastage and decrease the strength

In the graph, we can say that something directly related to the strength of the mixture increasing when the cement in the mixture is increased because theybut to get the best mixture cement should be between 300kg-m^3 to 400kg-m^3 more than this is not necessary but will be accepted but less than this will result in decreasing the strength

```python
sns.scatterplot(df['Cement'],df['Strength'])
```

### Blast furnace slag vs strength

In the case of blast furnace slag-fly ash blended systems, the reaction rate generally increases with higher amounts of slag and at higher activator amounts, but this can come at the cost of more challenging workability, including 

1. An increased amount of fly ash delayed the setting, reduced the compressive strength, modulus of elasticity

2. enhanced the homogeneity of the hardened matrix and limited the micro-cracking

from the below graph, we can say that the blast furnace slag is getting the strength max out as when it is 0 we can reach max strength also but if added will increase the rate of reaction and will increase the strength even more

```python
sns.scatterplot(df['Blast Furnace Slag'],df['Strength'])
```

### Fly ash vs Strength

Fly ash influenced both fresh and hardened binder matrix properties of cement-based on alkali-activated high-MgO blast furnace slag. Increasing the fly ash content delayed the initial setting time but had little effect on the final setting time, although the lag from initial to final setting remained much longer than is normal for Portland cement.

from the below graph, we can say that the Fly Ash is not much important in getting the strength max out as when it is 0 we are able to reach max strength also but if added should be between 100-125 because that can increase the rate of reaction

```python
sns.scatterplot(df['Fly Ash'],df['Strength'])
```

### Water vs Strength

Water plays an important part in making a mixture strong because if we increase the quantity of water it will result in making mixture waste and if decreases then could make the mixture wet and get dried out fast resulting loss in strength so according to the analysis we can say that water should be between 140 to 180 

```python
sns.scatterplot(df['Water'],df['Strength'])
```

### Superplasticizer vs strength

The addition of a naphthalene-based superplasticizer appeared to have very little effect on the initial and on the final setting times while a polycarboxylate-based superplasticizer showed an accelerating effect

from the below graph we can say that super plasticizer is not that neccesay in the making of mixture strength increase but if added should be between 8 kg-m^3 - 11 kg-m^3

```python
sns.scatterplot(df['Superplasticizer'],df['Strength'])
```

### coarse aggregate vs strength

As the sizes of coarse aggregate increases then the compressive strength will also increase with constant water to cement ratio

The workability of concrete get improved after increasing larger sizes of aggregates in concrete

from the below gaph we can say that coarse aggregate is important for making mixture strength increse and to do so we can add it up to 930  kg-m^3 - 980  kg-m^3

```python
sns.scatterplot(df['Coarse Aggregate'],df['Strength'])
```

### fine aggregate vs strength

Most natural sands from riverbeds or seashores are
smooth and rounded and are excellent aggregates.
Concrete made from it gave the highest compressive
strength at all ages followed by concrete made from
quarry dust, crushed palm kennel shell, saw-dust and
crushed groundnut shell respectively. These are due to
the fact that the better the surface texture, the higher the
compressive strength.

from the below graph we can say that fine aggregate are important as add them between 750 -800 could really increase the strength more then 80

```python
sns.scatterplot(df['Fine Aggregate'],df['Strength'])
```
